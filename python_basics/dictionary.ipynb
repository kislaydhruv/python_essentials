```python
### NESTED DICT ---> Keeping a dict inside the another dictionary 
```

```python
Nested_Dict={"Student_1":{"Name":"Amit", "Roll":21} , "Student_2":{"Name":"Amit", "Roll":43}}
```

```python
Nested_Dict
```

```python
# Iterate over these Nested Dict 
```

```python
for outer_key, Outer_value in Nested_Dict.items():
    print(f"Outer Key is :{outer_key} and outer value is : {Outer_value}")
    for key_1, value_1 in Outer_value.items():
        print(f"Inner Key is : {key_1} and Inner Value is : {value_1}")
```

```python
Nested_Dict.items()
```
