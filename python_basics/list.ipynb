### LIST METHODS

```python
#append() method is used to add an item to the end of the list.
list_item.append("banana")
print(list_item)
```

```python
# Insert() method is used to add an item at a specific index in the list.
list_item.insert(2,"amit")
print(list_item)
```

```python
# remove() method is used to remove the first occurrence of a specified value from the list.
list_item.remove("amit")
print(list_item)
```

```python
# pop() method is used to remove an item at a specific index in the list and return it.
popped_item = list_item.pop(3)
print(popped_item)
print(list_item)
```

```python
# index() method is used to find the index of the first occurrence of a specified value in the list.
index_of_item = list_item.index("singh")
print(index_of_item)
```

```python
list_item.append(9)
```

```python
list_item.append(8)
```

```python
# count() method is used to count the number of occurrences of a specified value in the list.
count_of_item = list_item.count(8)
print(count_of_item)
```

```python
#sort() method is used to sort the items of the list in ascending order.
list_item.sort()
print(list_item)
```

```python
# reverse() method is used to reverse the order of the items in the list.
list_item.reverse()
print(list_item)
```

```python
# clear() method is used to remove all items from the list.
list_item.clear()
print(list_item)
```

```python
# enumerate() method is used to add a counter to an iterable and return it as an enumerate object.
list_variable = ["Kislay", "Ranjan", 18, 5.6, True]
for index, item in enumerate(list_variable):
    print(index, item)  
```

```python
# len() method is used to return the number of items in the list.
length_of_list = len(list_variable)
print(length_of_list)
```

```python
list_itemss = ["apple", "banana", "cherry", "date", "elderberry"]
```

```python
length_list = [len(word) for word in list_itemss]
print(length_list)
```

```python
pair=[[i,j] for i in list_itemss for j in length_list]
print(pair)
```
