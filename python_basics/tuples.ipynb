### PACKING AND UNPACKING 

```python
# packing of tuples
packed_tuple=1, "hello", 3.14   
```

```python
type(packed_tuple)
```

```python
# unpacking of tuples
a, b, c = packed_tuple  
```

```python
type(a)
```

```python
print(a,b,c)
```

```python
type(b)
```

```python
# unpacking with a* operator
a, *b = packed_tuple
```

```python
print(b)
```

```python
type(b)
```
